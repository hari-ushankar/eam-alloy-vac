## Calculate vacancy migration energy in Cu using EAM/alloy

```python
import lammps_logfile as lmp_log
import numpy as np
from matplotlib import pyplot as plt
```

```python
from ase.build import bulk
```

```python
bCu = bulk("Cu", "fcc", a=3.615, cubic=True).repeat(3)
bCu.pop(0)
```

```python
from ase.io import write
bCu.write("bulkCu_small.lmp", format="lammps-data")
```

```python
bCu = bulk("Cu", "fcc", a=3.615, cubic=True).repeat(3)
bCu.pop(1)
bCu.write("bulkCu_small_1.lmp", format="lammps-data")
```

```python
energy = []
n_images = 9
for i in np.arange(0,n_images):
        energy.append(lmp_log.File("log.lammps.{}".format(i)).get("PotEng")[-1])
```

```python
plt.figure(dpi=400)
plt.plot(energy - energy[0],'-*')
plt.xlabel("Reaction coordinate")
plt.ylabel("Energy [eV]")
```
